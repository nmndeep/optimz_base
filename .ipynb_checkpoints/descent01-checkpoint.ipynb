```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Simple grad_descent

```python
def grad_des(start, iters, grad):
    points = [start]
    for ite in iters:
        points.append(points[-1]-ite*grad(points[-1]))
        
    return points
```

```python
def func1(x):
    ########  quadratic polynomial
    return 0.5*x.dot(x)
def grad(x):
    return x
```

```python

def graph_plt(val):
    plt.figure(figsize=(5, 5))
    plt.xlabel('Step')
    plt.ylabel('Error')
    plt.yscale('log')
    plt.plot(range(len(val)), val, linewidth=3.0)
```

```python
x0 = np.random.normal(0, 1, (1000))
```

```python
x_t = grad_des((x0),[0.2]*20, grad)
graph_plt([func1(x) for x in x_t])
plt.plot(range(len(x_t)), [func1(x0)]*len(x_t), label='Opt')
```

### Constrained

```python
############   Projection back to an affine subspace after every descent step
```

```python
####    Affine subspace
U = np.linalg.qr(np.random.normal(0, 1, (1000, 100)))[0]
a = np.random.normal(0, 1, 1000)

def project(x):
    return a + U.dot(U.T).dot(x-a)
```

```python
def proj_grad_des(start, iters, grad):
    points = [start]
    for ite in iters:
        points.append(project(points[-1]-ite*grad(points[-1])))
        
    return points
```

```python
x0 = np.random.normal(0, 1, (1000))
x_t = proj_grad_des(x0, [0.2]*20, grad)
```

```python
x_opt = project(0)
graph_plt([func1(x) for x in x_t])
plt.plot(range(len(x_t)), [func1(x_opt)]*len(x_t), label='Opt')
plt.legend()
```

### Least Squares with Tikhonov

```python
n, m = 200, 1000
alpha = 0.1
def least_sqt(A, b, x, alpha = 0.1):    
    return (0.5/n)*np.linalg.norm((A.dot(x)-b))**2 + (alpha/2)*x.dot(x)

def ls_grad(A, b, x, alpha = 0.1):
    return A.T.dot(A.dot(x)-b)/n + alpha*x
```

```python
A = np.random.normal(0, 1, (n, m))
b = A.dot(np.random.normal(0, 1, m))

###   Optimal x given using Monroe-Psuedo inverse
x_opt = np.linalg.inv(A.T.dot(A)+ alpha*np.eye(m)).dot(A.T).dot(b)
```

```python
x0 = np.random.normal(0, 1, m)
x_t = grad_des(x0, [0.1]*500, lambda x : ls_grad(A,b,x))
```

```python
obj = lambda x: least_sqt(A,b,x)
graph_plt([obj(x) for x in x_t])
plt.plot(range(len(x_t)), [least_sqt(A, b, x_opt)]*len(x_t), label='opt')
plt.legend()
```
